```python
import pandas as pd
```

```python
dfamz = pd.read_csv(r'Datasets\amazon_prime_titles-score.csv')
dfdsy = pd.read_csv(r'Datasets\disney_plus_titles-score.csv')
dfhu = pd.read_csv(r'Datasets\hulu_titles-score (2).csv')
dfnfx = pd.read_csv(r'Datasets\netflix_titles-score.csv')
```

```python
print(dfamz.shape)
dfamz.head()
```

```python
dfamz["id"] = 'a' + dfamz.show_id
```

```python
dfamz.rating.isnull().sum() #cantidad de nulos
```

```python
dfamz.rating.fillna('G', inplace=True)
```

```python
dfamz["date_added"] = pd.to_datetime(dfamz.date_added)
```

```python
dfamz[dfamz.select_dtypes(include = ['object']).columns] = dfamz[dfamz.select_dtypes(include = ['object']).columns].apply(lambda x: x.str.lower(), axis=1) #se reemplazan las columnas
```

```python
dfamz["duration_int"] = dfamz.duration.str.split(pat = " ", n = 1, expand = True).loc[:,0]
dfamz["duration_type"] = dfamz.duration.str.split(pat = " ", n = 1, expand = True).loc[:,1]
```
